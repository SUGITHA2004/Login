```python
# 3)Find the sum of first N natural numbers

def natural_numbers(N):
    total_sum = 0
    for i in range(1, N + 1):
        total_sum += i
    return total_sum

print(natural_numbers(20))





```

```python
# 4)Print muliplication of give no

num=int(input("Enter a number:"))
for i in range(1,num):
  print(num,"x",i,"=",num*i)
```

```python
# 5)factorial number

def fact(n):
    x = 1
    for i in range(1, n + 1):
        x = x * i
    return x


result = fact(6)
print(result)
```

```python
# 6) Fibonacci Series of N number

def fibonacci(n):
  a=0
  b=1
  if n==1:
    print(a)
  else:
    print(a)
    print(b)
    for i in range(2,n):
      c=a+b
      a=b
      b=c
      print(c)
fibonacci(10)
```

```python
# 7) count digit in a given no using loop

num=int(input("Enter a number:"))
count=0

if num==0:
  print("Its zero")
else:
  while num>0:
    num=num//10
    count+=1
print(count)

```

```python
# 8) Reverse a no using loop

num=int(input("Enter a number:"))
reverse=0

while num>0:
  remainder=num%10
  reverse=(reverse*10)+remainder
  num=num//10
print(reverse)
```

```python
# 9) The no Prime or not

num=int(input("Enter a number:"))
Is_Prime=True

for  i in range(2,num):
  if num%i==0:
    Is_Prime=False
    break
if Is_Prime:
  print("Its Prime")
else:
  print("Its not Prime")
```

```python
# 10) print the following pattern loop

# *
# * *
# * * *
# * * * *
# * * * * *

for i in range(1,10):
  for j in range(i):
    print("*",end=" ")
  print()

```
