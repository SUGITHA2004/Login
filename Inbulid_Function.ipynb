```python
## PRIME NUMBER

num = int(input("Enter a number: "))
is_prime = True

for i in range(2, num):
    if num % i == 0:
        is_prime = False
        break

if is_prime:
    print("It is prime")
else:
    print("It is not prime")
```

```python
## FIBNOACI SERIES

def fibonaci(x):
    a = 0
    b = 1

    if x == 1:
        print(a)
    else:
        print(a)
        print(b)
        for i in range(2, x):
            c = a + b
            a = b
            b = c
            print(c)

fibonaci(10)
```

```python
## FACTORIAL

def fact(n):
    x = 1
    for i in range(1, n + 1):
        x = x * i
    return x


result = fact(6)
print(result)
```
